```python
data=pd.DataFrame(df,columns=['overs','scores'])
```

```python
data
```

```python
from sklearn.cluster import KMeans
```

```python
# prompt: define kmeans

from sklearn.cluster import KMeans

kmean = KMeans(n_clusters=3)
```

```python
kmean.fit(data)
```

```python
centroids=kmean.cluster_centers_
```

```python
centroids
```

```python
plt.scatter(data['overs'],data['scores'],c=kmean.labels_.astype(float),s=50 ,alpha=1)
plt.scatter(centroids[:,0],centroids [:,1],c='red',s=50)
plt.xlabel('overs')
plt.ylabel('scores')
plt.show()
```
